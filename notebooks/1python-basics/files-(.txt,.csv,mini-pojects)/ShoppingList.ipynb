```python

class ShoppingList:
    """
    Shopping list manager that allows adding, removing, and saving items.
    """
    def __init__(self):
        self.items = []

    def add_item(self, item: str):
        self.items.append(item)

    def remove_item(self, item: str):
        if item in self.items:
            self.items.remove(item)

    def show_list(self):
        return "\n".join(self.items)

    def save_to_file(self, filename="shopping.txt"):
        with open(filename, "w", encoding="utf-8") as f:
            for item in self.items:
                f.write(item + "\n")

    def load_from_file(self, filename="shopping.txt"):
        try:
            with open(filename, "r", encoding="utf-8") as f:
                for line in f:
                    self.items = f.readlines()
        except FileNotFoundError:
            print("File does not exist")


# Demo
sl = ShoppingList()
sl.add_item("Milk")
sl.add_item("Bread")
sl.add_item("Eggs")
print("Shopping list:\n", sl.show_list())
sl.save_to_file()
```
